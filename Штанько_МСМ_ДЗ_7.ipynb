### **Задание 1:**

Выберем из общей таблицы те столбцы, с которыми мы будем работать (уберем REGION, т.к. можно использовать FEDERAL_OKRUG):

```python
X = data.drop(columns=['Q_OF_LIFE_INDEX', 'REGION'])
y = data['Q_OF_LIFE_INDEX']
X
```

Проведем все первоначальные действия с данными: возьмем Q_OF_LIFE_INDEX за целевую переменную и преобразуем качественную переменную в количественную с помощью OneHotEncoder и разделим данные на тестовую и обучающую выборки:

```python
X = pd.get_dummies(X, columns=['FEDERAL_OKRUG'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Сначала сделаем бэггинг (построим регрессию на индекс качества жизни):

```python
bagg = RandomForestRegressor(random_state=0,n_estimators=100)
bagg.fit(X_train, y_train)
bagg_pred = bagg.predict(X_test)
bagg_mse = mean_squared_error(y_test, bagg_pred)
bagg_rmse = np.sqrt(bagg_mse)

print(f'Качество бэггинга на тесте:\nMSE: {bagg_mse:.2f}\nRMSE: {bagg_rmse:.2f}')
```

Теперь сделаем то же самое для бустинга:

```python
grab = GradientBoostingRegressor(random_state=0, n_estimators=100)
grab.fit(X_train, y_train)
grab_pred = grab.predict(X_test)
grab_mse = mean_squared_error(y_test, grab_pred)
grab_rmse = np.sqrt(grab_mse)
print(f'Качество бустинга на тесте:\nMSE: {grab_mse:.2f}\nRMSE: {grab_rmse:.2f}')
```

```python
y_test.describe()
```

Таким образом, можно заметить, что градиентный бустинг справился лучше на тестовой выборке. Однако, в целом, модели показывают не очень хорошие результаты в силу того, что стандартное отклонение индекса счастья от среднего в тестовой выборке составляет 12, а модели в среднем ошибаются на 6 у.е.

### **Задание 2:**

```python
pokazateli = []
for i in np.arange(0.1, 1.1, 0.1):
    maxf = max(1, int(i * X_train.shape[1]))
    randfor = RandomForestRegressor(
        max_features=maxf,
        random_state=0,
        n_estimators=100
    )
    randfor.fit(X_train, y_train)
    predrand = randfor.predict(X_test)
    mserand = mean_squared_error(y_test, predrand)
    rmserand = np.sqrt(mserand)
    pokazateli.append({'maxik': maxf,'MSE': mserand, 'RMSE': rmserand})
```

```python
print('\n'.join([f"Кол-во признаков: {x['maxik']}, MSE: {x['MSE']:.2f}, RMSE: {x['RMSE']:.2f}" for x in sorted(pokazateli, key=lambda x: x['MSE'])]))
```

Получается, что лучше всего справляется модель, которая берет не более 9 признаков для разделения узла. Ее ошибка составляет 6.22 (по RMSE).

```python
best = sorted(pokazateli, key=lambda x: x['MSE'])[0]
print(f'У лучшей модели RandomForest, которую мы подобрали в предыдущем пункте, получились следующие результаты: \nMSE: {best["MSE"]:.2f}\nRMSE: {best["RMSE"]:.2f}')

print(f'\nКачество бэггинга на тесте получилось:\nMSE: {bagg_mse:.2f}\nRMSE: {bagg_rmse:.2f}')
print(f'\nКачество бустинга на тесте:\nMSE: {grab_mse:.2f}\nRMSE: {grab_rmse:.2f}')
```

Видно, что лучшая модель RandomForest справилась так же хорошо, как и бустинг, только в бустинге я не подбирал оптимальные параметры. Таким образом, вывод заключается в том, что модель здесь работают лучше, чем предсказывающее среднее, однако требуют доработки. К данным больше подошел бустинг, чем бэггинг, так как без настроек он выдает лучшие результаты. При этом, случайный лес смог догнать по качеству бустинг после настройки модели.
