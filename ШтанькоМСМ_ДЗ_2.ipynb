2. Между ВРП и долей женщин среди работников наблюдается отрицательная связь на уровне **-0.81**. Здесь скорее можно сделать вывод о том, что данный коэффициент корреляции показывает ложные результаты вследствие того, что рассматриваемые признаки могут зависеть и от других, что отражается в коэффициенте корреляции, поэтому здесь более грамотным выходом из этой ситуации был бы подсчет частного коэффициента корреляции. Проверим чистую зависимость между признаками с помощью избавления от зависимости от остальных признаков:


```python
import pingouin as pg
spisok = data.select_dtypes(exclude=['object']).columns.tolist()
spisok = list(filter(lambda x: x not in ('GRP', 'FEM_SHARE_IN_EMPLOYEES'), spisok))
part_corr = pg.partial_corr(data.select_dtypes(exclude=['object']), x='GRP', y='FEM_SHARE_IN_EMPLOYEES', covar=spisok,  method='pearson')
print(f'Коэффициент частной корреляции между ВРП и долей женщин среди работников за исключением влияния всех остальных\nпризнаков: {round(part_corr['r'].tolist()[0], 2)}')
```

Получилось, что корреляция скорректировалась аж на целых 0.6 по модулю. Теперь мы наблюдаем слабую линейную связь между этими признаками.

Теперь посмотрим на `матрицу p-значений статистического теста на значимость коэффициента корреляции для пар разных признаков`:

```python
import pandas as pd
import numpy as np
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr
pandas2ri.activate()
stats = importr('stats') 
psych = importr('psych') 

data_num = data.select_dtypes(exclude=['object'])
data_r = pandas2ri.py2rpy(data_num)

result = psych.corr_test(data_r, adjust="none", ci=True)
corr_matrix = result.rx2('r') 
p_matrix = result.rx2('p')     

p_matrix = pd.DataFrame(p_matrix, columns=data.corr(numeric_only=True).columns, index=data.corr(numeric_only=True).index)
p_matrix
```

Заметим, что при уровне значимости `0.05` большая часть гипотез отвергается, так как p-value < 0.05, что говорит о значимости коэффициентов корреляции. Результаты, как можно заметить, согласуются с матрицей корреляции, полученной выше. 
Теперь же для удобства визуализации `построим график p-value`:

```python
pgraph = sns.heatmap(p_matrix, cmap='Greens', annot=True, fmt='.2f', annot_kws={'size': 5})
pgraph.set_xticklabels(pgraph.get_xticklabels(), fontsize=7)
pgraph.set_yticklabels(pgraph.get_yticklabels(), fontsize=7)
pgraph.set_title('Матрица p-value', fontsize=18)
plt.show()
```

Если посмотреть на интересные значения p-value, то можно на пересечении темпа роста числа предприятий и рейтинга уровня жизни заметить его равным **0.98**. Таким образом, гипотеза не будет отвергаться на уровне значимости 0.05, следовательно, коэффициент корреляции незначим и равен 0. Если посмотреть на график корреляций, мы именно это и заметим: коэффициент корреляции для этих признаков равен **0**.

### **Задание 3:**

`Разделим женскую зарплату на две категории (высокая и низкая) по медиане` и `выведем на экран табличку`, которая отражает только необходимые нам признаки для построения корреляционного отношения:

```python
data['FEM_WAGE_RATE'] = pd.qcut(data['FEM_WAGE'], q=2, labels=['Низкая', 'Высокая'])
wages = data[['WAGE_GENDER_GAP', 'FEM_WAGE', 'FEM_WAGE_RATE']]
```

Теперь посчитаем `коэффициент корреляционного отношения` между признаками **WAGE_GENDER_GAP** и **FEM_WAGE**, который лежит в промежутке от нуля до единицы и показывает тесноту связи между признаками (направление связи не показывает). Считается с помощью деления на группы второго признака и подсчета межгрупповой вариации. Если межгрупповая вариация объясняет всю вариацию целевого признака, то коэффициент корреляционного отношения будет равен 1, 0 - наоборот. Посчитаем данный коэффициент:

```python
import pingouin as pg

anova_result = pg.anova(data=wages, dv='WAGE_GENDER_GAP', between='FEM_WAGE_RATE', detailed=True)
eta_squared = anova_result['np2'][0]

print(f"Коэффициент корреляционного отношения η между признаками равен {np.sqrt(eta_squared):.2f}")
print(f"Квадрат этого отношения равен {eta_squared:.2f}")
```

Квадрат корреляционного отношения между гендерным разрывом зарплат (Y) и зарплатами у женщин (X) показывает **долю вариации первого признака (Y), которую объясняет второй признак (X)**. Таким образом, `13% вариации гендерного разрыва зарплат объясняется вариацией признака X (зарплатами у женщин)`.
